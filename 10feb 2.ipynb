```python
#Selecting Data
import pandas as pd
data={
    "ID":[101,102,103,104,105],
    "Name":["Alice","Bob","Charlie","David","Eva"],
    "Age":[22,25,23,24,26],
    "Marks":[85,78,92,88,90]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
print("Displaying only the 'Names' Column\n")
print(df["Name"])
print("\nDisplaying only the 'Name' and 'Marks' Columns\n")
print(df[["Name","Marks"]])
print("\nLocation 0\n",df.iloc[0])
print("\nLocation 1 to 4\n",df.iloc[1:4])
print("\nLocation 1 to 3\n",df.iloc[0:3,1:3])
```

```python
# Insertion Sort
n = int(input("Enter number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

print("Original Array:", arr)

for i in range(1, n):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key

print("Sorted Array:", arr)

```

```python
# Binary Search
n = int(input("Enter the number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))
arr.sort()
print("\nSorted list:", arr)
key = int(input("\nEnter key element to be found: "))
low = 0
high = n - 1
found = False
while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print(f"Element found at position {mid + 1}")
        found = True
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1
if not found:
    print("Element not found")
```
