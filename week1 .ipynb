```python
#datasets
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
df.to_csv("iris_dataset.csv", index=False)
print("Dataset downloaded and saved successfully!")
print(df.head())
df["sepal_area"] = df["sepal_length"] * df["sepal_width"]
df.to_csv("iris_transformed.csv", index=False)
print("Dataset transformed and saved!")
```

```python
file_name = "sample_text.txt"
with open(file_name, "w") as file:
    file.write("Hello World\n")
    file.write("This is a Python file handling example\n")
    file.write("Counting lines words and characters")
print("File created and data written!")
with open(file_name, "r") as file:
    text = file.read()
lines = text.split("\n")
line_count = len(lines)
word_count = len(text.split())
char_count = len(text)
print("Number of lines:", line_count)
print("Number of words:", word_count)
print("Number of characters:", char_count)

```

```python
#list
a=[10,20,30]
print(type(a))
lst = [10, 20, 30, 40]
print("Initial list:", lst)
print("First element:", lst[0])
print("Last element:", lst[-1])
lst.append(50)
print("After append:", lst)
lst.insert(2, 25)
print("After insert:", lst)
lst.extend([60, 70])
print("After extend:", lst)
lst.remove(25)
print("After remove:", lst)
lst.pop()
print("After pop:", lst)
lst.pop(1)
print("After pop at index 1:", lst)
lst[0] = 100
print("After updating:", lst)
print("Index of 30:", lst.index(30))
print("Count of 20:", lst.count(20))

```

```python
# DICTIONARY OPERATIONS
d = {
    "name": "r sahasra",
    "age": 19,
    "course": "BTech"
}
print("Original dictionary:", d)
print("Name:", d["name"])
print("Age:", d.get("age"))
d["college"] = "XYZ University"
print("After adding:", d)
d["age"] = 21
print("After updating:", d)
d.pop("course")
print("After pop:", d)
removed_item = d.popitem()
print("After popitem:", d)
print("Removed item:", removed_item)
print("Keys:", d.keys())
print("Values:", d.values())
print("Items:", d.items())
print("Is 'name' a key?", "name" in d)
print("Length of dictionary:", len(d))
# Nested dictionary
student = {
    "id": 1,
    "info": {
        "name": "Amit",
        "marks": 70
    }
}
print("Nested dictionary value:", student["info"]["marks"])

# Clearing dictionary
d.clear()
print("After clear:", d)
```

```python
#set
fruits = {"apple", "banana", "cherry"}
print("Original Set:", fruits)
fruits.add("orange")
print("\nAfter add():", fruits)
fruits.update(["mango", "grapes"])
print("\nAfter update():", fruits)
fruits.remove("banana")
print("\nAfter remove('banana'):", fruits)
fruits.discard("pineapple")
print("\nAfter discard('pineapple'):", fruits)
removed_item = fruits.pop()
print("\nAfter pop():", fruits)
fruits_copy = fruits.copy()
print("\nCopied Set:", fruits_copy)
print("\nLength of set:", len(fruits))
# Checking membership
if "apple" in fruits:
    print("\n'apple' is present in the set")
A={1,2,3,4,5}
B={4,5,6,7,8}
print("\nUnion:", A.union(B))
print("Intersection:", A.intersection(B))
print("Difference (A - B):", A.difference(B))
print("Symmetric Difference:", A.symmetric_difference(B))
print("\nIs A subset of B?", A.issubset(B))
print("Is A superset of B?", A.issuperset(B))
fruits.clear()
print("\nAfter clear():", fruits)
```
