```python
#32 Implement the lasso model
from sklearn.linear_model import Lasso
import numpy as np
x=np.array([[1,25,160],[2,30,165],[3,35,170]])
y=np.array([30000,40000,50000])
model=Lasso(alpha=0.1)
model.fit(x,y)
print(model.coef_)
```

```python
#33 Implement the similarity and distances for any dataset
from scipy.spatial import distance
from scipy.stats import pearsonr
import numpy as np
p1=np.array([2,3])
p2=np.array([5,7])
euc=distance.euclidean(p1,p2)
mh=distance.cityblock(p1,p2)
cb=distance.chebyshev(p1,p2)
hm=distance.cosine(p1,p2)
cs=1-cb
set1,set2=set(p1),set(p2)
js=len(set1&set2)|len(set1&set2)
r=pearsonr(p1,p2)
print("Euclidean Distance",euc)
print("Manhatten Distance",mh)
print("Chebyshev Distance",cb)
print("Cosine Similarity",cs)
print("Jaccord Distance" ,js)
print("Pearsorns coefficient",r)
```

```python
#34 Write a code to implement the K-means clustering(Unsupervised learning) algorithm
import numpy as np
A=np.array([[2,3],[3,4],[4,2]])
B=np.array([[8,7],[9,6],[10,8]])
P=np.array([5,5])
cA=A.mean(axis=0)
cB=B.mean(axis=0)
dA=np.linalg.norm(P-cA)
dB=np.linalg.norm(P-cB)
if dA<dB:
  print("Point belongs to cluster A")
else:
  print("Point belongs to cluster B")
```
